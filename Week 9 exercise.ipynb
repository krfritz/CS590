```r
Bparent = function(x){
    if (grepl('Bailey', x)){
   return('yes') 
}else{
    return('no')
    }
}
lines['Bailey_Parent'] = sapply (lines[['Parentage']], Bparent)
head(lines)
```

```r
nrow(lines[lines['Bailey_Parent'] == 'yes',])
```

```r
percent = round(nrow(lines[lines['Bailey_Parent'] == 'yes' & lines['FAG'] == 'ol ol',]) / nrow(lines) * 100, 1)
noquote(sprintf('%s%%', percent))
```

```r
lines['Parentage'] = gsub ('Bailey', 'N03081T', lines[['Parentage']])
head(lines, 10)
```

```r
yields = read.csv("peanut_yield.txt", sep='\t', header = TRUE)
yields
```

```r
peanut_db = merge(lines, yields, by.x='NC_Accession', by.y='NC_Accession', all.y=TRUE)
```

```r
mean(peanut_db[peanut_db['NC_Accession'] != 'Bailey',]$Yield, na.rm = TRUE)
```

```r
peanut_db[peanut_db['NC_Accession'] != 'Bailey',]
```

```r
peanut_db <- peanut_db[peanut_db['NC_Accession'] == 'Bailey' | peanut_db['Bailey_Parent'] == 'yes',]
Line <- c('Bailey', 'Other')

Yield <- c(mean(peanut_db[peanut_db['NC_Accession'] == 'Bailey',]$Yield, na.rm=TRUE), mean(peanut_db[peanut_db['NC_Accession'] != 'Bailey',]$Yield, na.rm=TRUE))
dframe <- data.frame(row.names = Line, Yield)
dframe
```
