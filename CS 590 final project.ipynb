```r
#Make a vector with the parents of my population
vector = c('PI371521', 'PI471952', 'PI493693')
```

```r
#create a function with the vector for whether or not it is one of the parents
graph_condition <- function(x){
    if (x %in% vector){
        return('parent of FRT')
    }else{
        return('not parent of FRT')
    }
}
graph_condition(vector)
```

```r
#Create a new column with whether or not it is a parent
phenotypes$parents = sapply(phenotypes[['PI_No']], graph_condition)
phenotypes
```

```r
#Add an asthetic so that it colors the daata points based on whether or not it is a parent for my FRT population
qplot(PI_No, folic, data=phenotypes, color=parents)
```
